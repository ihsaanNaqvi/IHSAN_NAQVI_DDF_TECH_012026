```python
import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime, timedelta
import random
```

SETUP

```python
fake = Faker()

NUM_RECORDS = 150_000

categories = [
    "Electronics", "Phone Accessories", "Home Appliances",
    "Fashion", "Sports", "Books"
]

regions = [
    "São Paulo", "Rio de Janeiro", "Minas Gerais",
    "Paraná", "Bahia", "Rio Grande do Sul"
]
```

HELPER FUNCTIONS

```python
def random_date(start, end):
    delta = end - start
    return start + timedelta(days=random.randint(0, delta.days))


def generate_product_description(category):
    materials = ["Premium Plastic", "Aluminum", "PU Leather", "Cotton", "Steel"]
    features = [
        "durable", "lightweight", "eco-friendly",
        "high performance", "ergonomic design"
    ]
    return (
        f"This {category.lower()} product is made of "
        f"{random.choice(materials)} and offers "
        f"{random.choice(features)} features. "
        f"Ideal for daily use and long-term reliability."
    )
```

DATA GENERATION

```python
data = []

start_date = datetime(2021, 1, 1)
end_date = datetime(2025, 1, 1)

for i in range(NUM_RECORDS):
    category = random.choice(categories)
    price = round(random.uniform(5, 500), 2)
    quantity = random.randint(1, 5)

    data.append({
        "product_id": f"P{i:07d}",
        "title": fake.sentence(nb_words=6),
        "description": generate_product_description(category),
        "category": category,
        "price": price,
        "quantity": quantity,
        "order_date": random_date(start_date, end_date),
        "region": random.choice(regions)
    })

df = pd.DataFrame(data)
```
